# Pre-processing

```
#import and install packages and tools
from pandas.compat import StringIO
from collections import Counter
import itertools
import numpy as np
import string
import re
from string import punctuation
!pip install --upgrade pandas
```

```
#Lower case, remove punctuation, remove single letters, remove spaces
train_lower = train['Phrase'].apply(lambda x: x.lower())
train_clean = train_lower.apply(lambda x: x.translate(None, string.punctuation))
train_clean = pd.DataFrame(train_clean)
train_clean.columns = ['clean_phrase']
train_clean['clean_phrase'] = train_clean['clean_phrase'].str.replace(r"\b[a-zA-Z]\b", "")
train_clean['clean_phrase'] = train_clean['clean_phrase'].str.strip()
print train_clean.head(10)
type(train_clean)
type(train)
```

```
train = pd.DataFrame(train)
train_clean = pd.DataFrame(train_clean)
train_cleaner = pd.concat([train, train_clean], axis=1)
train_cleaner[['PhraseId','SentenceId','Phrase','clean_phrase','Sentiment']].loc[train_cleaner.SentenceId == 3]
```

```
#Count words per phrase + sentiment (0 - negative, 1 - somewhat negative, 2 - neutral, 3 - somewhat positive, 4 - positive)
train_cleaner['word_count'] = train_cleaner['clean_phrase'].apply(lambda x: len(str(x).split()))
train_cleaner[['SentenceId','clean_phrase','word_count','Sentiment']].loc[train_cleaner.SentenceId == 2]
```

```
#Calculating vocabulary size, word count, top words
tok_indx = dict()
unigram_counts = Counter()
clean_phrases = train_cleaner['clean_phrase'].tolist()
clean_phrases = [word.split() for word in clean_phrases]
for ii, clean_phrase in enumerate(clean_phrases):
    for token in clean_phrase:
        unigram_counts[token] += 1
        if token not in tok_indx:
            tok_indx[token] = len(tok_indx)
indx_tok = {indx:tok for tok,indx in tok_indx.items()}
print('vocabulary size: {}'.format(len(unigram_counts)))
#labels = ['Words' , 'Counts']
df = pd.DataFrame.from_dict(unigram_counts, orient='index')
df.columns = ['Counts']
df= df.sort_values(ascending=False, by=['Counts'])
#df['Word'] = unigram_counts.keys()
#df['Counts'] = unigram_counts.values()
df
#print('most common: {}'.format(unigram_counts.most_common(100)))
```

```
from stanfordcorenlp import StanfordCoreNLP
from nltk.parse.stanford import StanfordDependencyParser
from graphviz import Source
import os
import pydot
os.environ['CLASSPATH'] = r'stanford-parser-full-2018-02-27/'

nlp = StanfordCoreNLP('/content/stanford-corenlp-full-2018-01-31/')
sdp = StanfordDependencyParser()
#sdp = StanfordDependencyParser('/content/stanford-parser-full-2018-02-27/stanford-parser.jar')

sentence = 'A series of escapades demonstrating the adage that what is good for the goose is also good for the gander , some of which occasionally amuses but none of which amounts to much of a story .'
print 'Tokenize:', nlp.word_tokenize(sentence)
print 'Part of Speech:', nlp.pos_tag(sentence)
print 'Named Entities:', nlp.ner(sentence)
print 'Constituency Parsing:', nlp.parse(sentence)
print 'Dependency Parsing:', nlp.dependency_parse(sentence)

result = list(sdp.raw_parse(sentence))

dep_tree_dot_repr = [parse for parse in result][0].to_dot()
source = Source(dep_tree_dot_repr, filename="dep_tree", format="png")


nlp.close()
```
